```python
#importing varies libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#here read the dataset
df = pd.read_csv('walmart.csv')
```

```python
#show the entire dataset
df
```

```python
#1. Understanding User Demographics:
#	Distribution of Users: Analyze the distribution of users across Age groups, Gender, City Category, and Marital Status.
#Use histograms or bar charts to visualize the spread.

#here below is the count in dataframe of various age groups
sns.histplot(data = df, x = "Age")
plt.title("Distribution of the Age")
plt.show()
```

```python
sns.histplot(data = df, x = "Gender")
plt.title("Distribution of the Gender")
plt.show()
```

```python
sns.histplot(data = df, x = "City_Category")
plt.title("Distribution of the City_Category")
plt.show()
```

```python
sns.histplot(data = df, x = "Marital_Status")
plt.title("Distribution of the Marital_Status")
plt.show()
```

```python
#User Age and Purchase: Investigate how purchase amount varies across Age groups. Consider boxplots or scatter plots to identify trends.

sns.boxplot(x="Age", y="Purchase", data=df)
plt.title("Distribution of How purchase amount varies across Age groups")
plt.show()
```

```python
#Gender and Product Category: Explore if there are any significant differences in the Product Category purchased between genders.
#You can use stacked bar charts or compare means.

#Use pd.crosstab to create a table showing the count of each product category by gender.
crosstab = pd.crosstab(df['Product_Category'], df['Gender'])
```

```python
crosstab.plot(kind='bar', stacked=True,  color=['skyblue', 'lightgreen'])

plt.title('Product Category by Gender')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
#Marital Status and Purchase: Analyze if Marital Status has any influence on the amount spent.
#This could be visualized with boxplots or scatter plots with Purchase amount on the Y-axis.

sns.boxplot(x="Marital_Status", y="Purchase", data=df)
plt.title("Distribution of Marital Status V/S Purchase")
plt.show()
```

```python
#2. User Location and Purchase Behavior:
#City Category and Purchase: See if there are spending differences between City Category groups (A, B, C).
#You can use boxplots or ANOVA tests to compare means.


sns.boxplot(x="City_Category", y="Purchase", data=df)
plt.title("Distribution of City_Category V/S Purchase")
plt.show()
```

```python
#Length of Stay and Purchase: Investigate if the number of years a user has stayed in their current city
 #(StayInCurrentCityYears) impacts purchase amount. Utilize scatter plots or correlation analysis.

sns.scatterplot(x="Stay_In_Current_City_Years", y="Purchase", data=df)
plt.title("Distribution of Stay_In_Current_City_Years V/S Purchase")
plt.show()
```

```python
#3. User Purchase Behavior:
#Purchase Distribution: Analyze the distribution of purchase amounts. Histograms are useful to identify patterns and potential outliers.


sns.histplot(data = df, x = "Purchase")
plt.title("Distribution of the Purchase")
plt.show()
```

```python
#Frequency of Purchase: If there's a time component in the data, use it to calculate purchase frequency per user and explore its distribution.

sns.histplot(data = df, x = "Product_Category")
plt.title("Distribution of the Product_Category")
plt.show()
```

```python
#4. Product Category Analysis:
#Distribution of Product Categories: Identify the most frequently purchased product categories. Use bar charts or pie charts for visualization.


# Calculate the frequency of each product category
product_category_counts = df['Product_Category'].value_counts()
```

```python
plt.pie(product_category_counts, labels=product_category_counts.index, autopct='%1.1f%%', startangle=140,
        colors=plt.cm.Paired(range(len(product_category_counts))))
# Adding a title
plt.title('Distribution of Most Frequently Purchased Product Categories')

# Display the pie chart
plt.show()
```

```python
#Purchase by Category and Age: Explore if there are interactions between Age groups and Product Category preferences.
#Consider using stacked bar charts or crosstabulation.
# Create a pivot table
pivot_table = df.pivot_table(index='Age', columns='Product_Category', aggfunc='size', fill_value=0)
```

```python
# Plot stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(13, 7))

plt.title('Product Category Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Product Category')
plt.show()
```
