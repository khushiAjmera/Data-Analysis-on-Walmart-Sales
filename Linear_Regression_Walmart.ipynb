```python
#Predictive Modeling: Build models to predict purchase amount based on user characteristics or product categories.
#This could involve techniques like linear regression or decision trees.

#simple linear Regression
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df=pd.read_csv("walmart.csv")
df
```

```python
x = df["Product_Category"]
y=df["Purchase"]
plt.scatter(x,y)
plt.xlabel("Product_Category")
plt.ylabel("Purchase")
```

```python
from sklearn.linear_model import LinearRegression
x=df["Product_Category"].values.reshape(-1,1)
y=df["Purchase"]
lr_model=LinearRegression()
lr_model.fit(x,y)
y_pred=lr_model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.xlabel("Product_Category")
plt.ylabel("Purchase")
plt.plot(x,y_pred)
```
